# **Q.3) Does gender affect who gets searched during a stop ?**

Calculating the number of searches conducted.

```python
df.search_conducted.value_counts()
```

Finding the count of searches for Male and Female

```python
df.groupby('driver_gender').search_conducted.sum()
```

From this, we can conclude that the Male drivers are most likely to be checked by the policemen at the checkposts.

# **Q.4) What is the mean stop_duration ?**

The stop_duration column has values as follows.

```python
df.stop_duration.value_counts()
```

Since these values are not integers, mean for these values couldn't be calculated. so that we have to convert them into integers by taking mean for those time period.

```python
df['stop_duration']=df['stop_duration'].map({'0-15 Min':7.5 , '16-30 Min':24 , '30+ Min':45 ,})
```

Here the mean for each value has been updated. For 30+ mins, the value ranges from 30-60 mins, mean has been calculated as 45.

gettting the dataset to check whether the column has been updated.

```python
df
```

Calcuting mean using mean() function

```python
df['stop_duration'].mean()
```
